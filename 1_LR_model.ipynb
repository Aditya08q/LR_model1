# Quality

```python
df['Price ($)'].unique()
```

```python
df['Price ($)'] = df['Price ($)'].str.replace('$', '', regex=False) \
                                 .str.replace(',', '', regex=False) \
                                 .str.strip() \
                                 .astype(int)

```

```python
df['Storage '].unique()
```

```python
df['Storage '].str.split(' ').str.get(0)
```

```python
# Then clean the 'Storage' column
df['Storage '] = df['Storage '].str.replace('GB', '', regex=False).str.strip().astype(int)
```

```python
df.info()
```

```python
# Then clean the 'Storage' column
df['RAM '] = df['RAM '].str.replace('GB', '', regex=False).str.strip().astype(int)
```

```python
df['Camera (MP)'].unique()
```

```python
import re

def sum_camera_mp(value):
    # Extract all numbers and convert to integers, then sum
    numbers = map(int, re.findall(r'\d+', value))
    return sum(numbers)

df['Camera (MP)'] = df['Camera (MP)'].apply(sum_camera_mp)
```

```python
df['Camera (MP)'].unique()
```

```python
df['Screen Size (inches)'].unique()
```

```python
import re

def extract_main_screen_size(value):
    value = str(value)
    # Find the first decimal number in the string
    match = re.search(r'\d+\.?\d*', value)
    if match:
        return int(float(match.group()))
    return None  # or return 0 if you prefer

df['Screen Size (inches)'] = df['Screen Size (inches)'].apply(extract_main_screen_size)
```

# Model

```python
x=df.drop(columns = 'Price ($)')
y = df['Price ($)']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(x[['Brand','Model','Storage ','RAM ','Screen Size (inches)','Camera (MP)','Battery Capacity (mAh)']])
```

```python
ohe.categories_
```

```python
categorical_features = ['Brand', 'Model'] 
```

```python
numerical_features = ['Storage ', 'RAM ', 'Screen Size (inches)', 'Camera (MP)', 'Battery Capacity (mAh)']
```

```python
column_trans = make_column_transformer(
    (OneHotEncoder(handle_unknown='ignore'), ['Brand', 'Model']),
    remainder='passthrough'  # This keeps numerical columns
)
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,lr)
```

```python
pipe.fit(xtrain,ytrain)
```

```python
y_pred = pipe.predict(xtest) 
```

```python
r2_score(ytest,y_pred)
```

```python
scores =[]
for i in range (1000):
    xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2,random_state =1)
    lr = LinearRegression()
    pipe = make_pipeline(column_trans,lr)
    pipe.fit(xtrain,ytrain)
    y_pred = pipe.predict(xtest)
    scores.append(r2_score(ytest,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2,random_state = np.argmax(scores))
lr = LinearRegression()
pipe = make_pipeline(column_trans,lr)
pipe.fit(xtrain,ytrain)
y_pred = pipe.predict(xtest)
r2_score(ytest,y_pred)
```
